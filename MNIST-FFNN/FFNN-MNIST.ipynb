```python
import torch
import torch.nn.functional as F
import torchvision

import numpy as np
import matplotlib.pyplot as plt
```

Load the train and test dataset into DataLoader

```python
batch_size_train = 64
batch_size_test = 128
```

```python
train_loader = torch.utils.data.DataLoader(
  torchvision.datasets.MNIST('dataset/', train=True, download=True,
                             transform=torchvision.transforms.Compose([
                               torchvision.transforms.ToTensor(),
                               torchvision.transforms.Normalize(
                                 (0.1307,), (0.3081,))
                             ])),
  batch_size=batch_size_train, shuffle=True)

test_loader = torch.utils.data.DataLoader(
  torchvision.datasets.MNIST('dataset/', train=False, download=True,
                             transform=torchvision.transforms.Compose([
                               torchvision.transforms.ToTensor(),
                               torchvision.transforms.Normalize(
                                 (0.1307,), (0.3081,))
                             ])),
  batch_size=batch_size_test, shuffle=True)
```

Sample examples of the training dataset

```python
image, label = next(iter(train_loader))
```

```python
num = 10
num_row = 2
num_col = 5
```

```python
fig, axes = plt.subplots(num_row, num_col, figsize=(2.5*num_col, 3*num_row))
for i in range(num):
    ax = axes[i//num_col, i%num_col]
    ax.imshow(image[i].reshape(28,28), cmap='gray')
    ax.set_title(f"Ground truth: {label[i]}")
plt.tight_layout()
plt.show()
```
